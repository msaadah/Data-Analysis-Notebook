```python
clim_data = read_csv('../Desktop/cmiday18.txt', delimiter='\t')
display(HTML(clim_data.to_html()))
```

```python
new_clim_data = clim_data[clim_data.columns[~clim_data.columns.str.endswith('er')]]
print len(new_clim_data)
```
