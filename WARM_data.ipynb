## Cleaning Data

```python
clim_data.tail(10)
```

```python
#delete NaNs

new_clim_data = clim_data[clim_data.columns[~clim_data.columns.str.endswith('er')]]  # removes columns with NaN
new_clim_data.drop(new_clim_data.tail(9).index,inplace=True)                         # drop last n rows
new_clim_data.drop(new_clim_data.head(1).index,inplace=True)                         # drop first n rows
del new_clim_data['site']
```

```python
#convert value types from object to float

new_clim_data = new_clim_data.replace('---', 0.0)
for col in new_clim_data.columns:
    new_clim_data[col] = new_clim_data[col].astype(float)
```

```python
display(HTML(new_clim_data.to_html()))
```

```python
new_clim_data.info()
```

```python
group = new_clim_data.groupby(['year'])
for key,item in group:
    display(HTML(group.get_group(key).to_html()))
```
