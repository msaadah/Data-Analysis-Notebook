```python
labels = ['Frogs', "Hogs", "dogs", "logs"]
fracs = [15,30,45,10]

n=14
h=2
w=h*n

fig,axs = plt.subplots(2,7)

fig.set_size_inches(w,h)

for pk in range(n):
    axs[pk].pie(fracs, labels = labels, autopct = '%1.1f%%', shadow = True)
    
plt.show()
    
```
