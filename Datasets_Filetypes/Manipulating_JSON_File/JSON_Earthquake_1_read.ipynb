Get details from: https://earthquake.usgs.gov/data/comcat/data-eventterms.php#mag

```python
url ='https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2018-01-01&minmagnitude=3.1'
page = requests.get(url)
urlData= page.content
```

```python
parsed_json = json.loads(urlData)
```

```python
columns=[]
list_of_list = []
first_time=1
for key in parsed_json['features']:
    data = []
    for k in key['properties']:
        if(first_time):
            columns.append(k)
        data.append(key['properties'][k])
    first_time=0
    list_of_list.append(data)
```

```python
data = pd.DataFrame(list_of_list, columns=columns)
data = data.set_index('code')
```

```python
for i in data.columns:
    if data[i].dtype == 'object':        
        data[i] = data[i].str.strip(',')
    else:
        data[i] = data[i].astype(float)
```
