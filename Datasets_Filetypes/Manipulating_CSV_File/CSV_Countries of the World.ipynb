# 4. Cleaning Data into Readable Format  

***
    
## A.   CSV cleaning

**This file needs different column titles to eventually change category types to either category or float.**

```python
df.columns = ['Country','Region', 'Population', 'Area', 'Density', 'Coastline', 'Migration', 'Infant_Mortality', 'GDP', 'Literacy', 'Phones', 'Arable', 'Crops', 'Other', 'Climate', 'Birthrate', 'Deathrate', 'Agriculture', 'Industry', 'Service']


df.Population = df.Population.astype(str).replace(",",".").astype(float)
df.Area = df.Area.astype(str).replace(",",".").astype(float)
df.Density = df.Density.astype(str).str.replace(",",".").astype(float)
df.Coastline = df.Coastline.astype(str).str.replace(",",".").astype(float)
df.Migration = df.Migration.astype(str).str.replace(",",".").astype(float)
df.Infant_Mortality = df.Infant_Mortality.astype(str).str.replace(",",".").astype(float)
df.GDP = df.GDP.astype(str).str.replace(",",".").astype(float)
df.Literacy = df.Literacy.astype(str).str.replace(",",".").astype(float)
df.Phones = df.Phones.astype(str).str.replace(",",".").astype(float)
df.Arable = df.Arable.astype(str).str.replace(",",".").astype(float)
df.Crops = df.Crops.astype(str).str.replace(",",".").astype(float)
df.Other = df.Other.astype(str).str.replace(",",".").astype(float)
df.Climate = df.Climate.astype(str).str.replace(",",".").astype(float)
df.Birthrate = df.Birthrate.astype(str).str.replace(",",".").astype(float)
df.Deathrate = df.Deathrate.astype(str).str.replace(",",".").astype(float)
df.Agriculture = df.Agriculture.astype(str).str.replace(",",".").astype(float)
df.Industry = df.Industry.astype(str).str.replace(",",".").astype(float)
df.Service = df.Service.astype(str).str.replace(",",".").astype(float)


df.Country = df.Country.astype('category')
df.Region = df.Region.astype('category')


df.head()
```

**Histogram to assess resting blood pressure distribution.**

```python
x = df['Density']
num_bins = 5
n, bins, patches = plt.hist(x, num_bins, alpha=0.5)
plt.title('Density', fontweight='bold')
plt.xlabel('Density')
plt.ylabel('Count')
plt.savefig('images/histogram.png')
plt.show()
```

**Create heatmap using .corr() to visualize correlation of all categories. Diagonal should all be 1.0 because comparing against itself.**

## H: Box Plot

```python
trace0 = go.Box(
    y = df.Birthrate,
    name = "Birthrates",
    boxpoints='all',
    pointpos=0,
    fillcolor='rgba(255, 255, 255, 0.8)',
    marker = dict(color = 'rgba(0, 255, 0, 0.8)')
    )

trace1 = go.Box(
    y = df.Deathrate,
    boxpoints='all',
    name = "Deathrates",
    pointpos=0,
    fillcolor='rgba(255, 255, 255, 0.8)',
    marker = dict(color = 'rgba(255, 0, 0, 0.8)')
    )

trace2 = go.Box(
    y = df.Infant_Mortality,
    boxpoints='all',
    name = "Infant Mortality",
    pointpos=0,
    fillcolor='rgba(255, 255, 255, 0.8)',
    marker = dict(color = 'rgba(0, 0, 255, 0.8)')
    )

plotdata = go.Figure(data = [trace0,trace1,trace2])
iplot(plotdata)
#plt.savefig('box_plot.png', bbox_inches="tight")
```

## J: World Map Projection

```python
def world_map_projection(col):
    dictdata = dict(type='choropleth', locations = df.Country,
    locationmode = 'country names', z = df[col],
    text = df.Country, colorbar = {'title':col})

    layout = dict(title=col+' Projections',
    geo = dict(projection={'type':'natural earth'}))

    choromap = go.Figure(data = [dictdata],layout = layout)

    iplot(choromap)
```

```python
world_map_projection('GDP')
world_map_projection('Phones')
world_map_projection('Literacy')
world_map_projection('Climate')
```
