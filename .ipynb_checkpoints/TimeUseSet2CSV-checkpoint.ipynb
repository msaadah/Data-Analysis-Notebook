```python
#CONVERT DATA TIMES TO DECIMAL VALUES

for i in df.columns:
    if i=='SEX' or i=='GEO/ACL00' or i=='Total':
        df[i]=df[i].str.strip().astype('category')
    else:
        #df[i] = df[i].str.replace(':', '.') 
        print(df[i])
        #replaces : with . in data
        print ('--------------------BEFORE---------------------')
        for j in range(len(df[i])):                                 #sets empty cells to 0.00; if prints FOUND then problem
            if df.iloc[j][i]==':':
                df.set_value(j,i,"00:00")                             
        df[i]=df[i].astype(str)+':00'

        print(df[i])
    
        print ('--------------------------MIDDLE----------------')
        #df[i]=df[i].astype(float)
        df[i]=pd.to_timedelta(df[i],'s')
        df[i]=df[i].dt.seconds/3600.0
        df[i]=df[i].round(2)
        #converts data to float type to keep numerical value
        #df[i]=df[i].np.timedelta64()
        #df[i] = df[i].apply(lambda x: x / np.timedelta64(1, 'h'))
    
        print(df[i])
        print ('-------------------AFTER------------------')
            

```
