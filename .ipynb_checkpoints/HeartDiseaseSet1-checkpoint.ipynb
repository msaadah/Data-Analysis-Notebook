```python
df = pd.read_csv('Desktop/Data/cleveland_processed.csv', header = None, index_col = False, delimiter = ',')
df.columns = ['Age','Sex','Chest_Pain','RestBP','Chol','FBS', 'RestECG', 'MaxHR', 'Exang', 'Oldpeak', 'Slope', 'Ca', 'Thal', 'Num']

print (df.head())
```

```python
df.plot()
plt.title('Plot of All Data', fontweight='bold')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)    #place legend box to right of plot
plt.show()
```

```python
df.plot()

plt.yscale('log')                                                 # makes plot log scale 

plt.title('Log Plot of All Data', fontweight='bold')
plt.xlabel('Index')
plt.ylabel('Log Scale')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) 
plt.show()
```

```python
age_pos = np.arange(len(df['Age']))
plt.bar(age_pos,df['Age'])
plt.xticks(age_pos)
plt.show()
```

```python
#Age distribution of study
x = df['Age']
num_bins = 50
n, bins, patches = plt.hist(x, num_bins, alpha=0.5)
plt.title('Age', fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Log Scale')
plt.show()
```

```python
#normalize distribution
age_hist = np.random.normal(size = 1000)
plt.hist(age_hist, normed=True, bins=30)
plt.ylabel('Age')
plt.show()
```
